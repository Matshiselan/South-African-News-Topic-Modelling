## USD/ZAR Exchange Rate
This script fetches the exchange rates between the US Dollar (USD) and the South African Rand (ZAR) between the start and end dates of the articles

```python
import pandas as pd
import requests
```

```python
articles_df = pd.read_csv('data/articles.csv')
articles_df.head()
```

```python
pip install yfinance
```

```python
import yfinance as yf

articles_df.date = pd.to_datetime(articles_df.date)
start_date = articles_df.date.min().strftime('%Y-%m-%d')
end_date = articles_df.date.max().strftime('%Y-%m-%d')

# Fetch USD/ZAR historical rates using yfinance
ticker = 'USDZAR=X'
zar_data = yf.download(ticker, start=start_date, end=end_date)
zar_rates = zar_data[['Close']].rename(columns={'Close': 'ZAR'})
zar_rates.head()
```

```python
zar_rates.to_csv('data/zar_rates.csv')
```
